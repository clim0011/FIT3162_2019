# Latent Dirichlet allocation (LDA)

Import libraries and open file

```python
import pandas as pd
import re
import nltk
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer
from gensim import corpora, models

# Read file
file1 = pd.read_csv("all_PosNeg.csv")
```

Funtion use to preprocess data

```python
# Lemmatize a string
def lemmatize(text):
    # Lemmatize with verb
    return WordNetLemmatizer().lemmatize(text, pos='v')

# Preprocess
def preprocess(text):
    result = []
    # Make text to lower case
    for token in gensim.utils.simple_preprocess(text):
        # Remove Stopwords and word that is shorter than 3 character
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize(token))
    return result
```

```python
# Preprocess negative review
processed_docs = file1['Positive_Review'].map(preprocess)

# Make the list to dictionary type
dictionary = gensim.corpora.Dictionary(processed_docs)

# Filter the dictionary that has frequency more than 2 and do not appear more than 50% of the total extracted word count and keep the most 100000 item
dictionary.filter_extremes(no_below=2, no_above=0.5, keep_n=100000)

# Convert the text into bag of word format (token id, token count)
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

# TF-IDF the corpus
tfidf = models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]

# LDA the corpus
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, num_topics=2, id2word=dictionary, passes=2, workers=4)
```

Write and print the topic into a file

```python
file_topic = open("topic.txt","w")

# print topic
for idx, topic in lda_model_tfidf.print_topics(-1):
    print('\nTopic: {} \nWord: {}'.format(idx, topic))
    file_topic.write('\nTopic: {} \nWord: {}'.format(idx, topic))
file_topic.close()
```

Calculate the LDA mark for all the review then write into a file

```python
# Find lda mark for all sentences respectively to business and leisure and sort 
array = []
for i in range(len(bow_corpus)):
    lst = lda_model_tfidf[bow_corpus[i]]
    lst.sort(key=lambda tup: tup[1])
    lst.reverse()
    array.append(lst)
    
# Write all lda mark into text file
file1 = open("result.txt","w")
index = 0
for item in array:
    file1.write(str(index) + '\n')
    for data in item:
        file1.write(' '.join(str(s) for s in data) + '\n')
    file1.write('_____ _________ _________' + '\n')
    index = index + 1
file1.close()

# Filter the lda mark
file_arr = open("result.txt","r")
file_remove = open("final_result.txt","w")
lines = file_arr.readlines()

array_a = []
total = 0

for line in lines:
    line = line.rstrip()
    line = line.split(' ')
    if (len(line) == 1):
        tup = (line)
    if (len(line) == 2 and total < 0.7):
        total = total + float(line[1])
        tup.append(line)
    if (len(line) == 3):
        total = 0
        array_a.append(tup)

for item in array_a:
    file_remove.write(item[0] + '\n')
    for data in item:
        if (isinstance(data,list)):
            file_remove.write(' '.join(str(s) for s in data) + '\n')
    file_remove.write('_____ _________ _________' + '\n')
file_arr.close()
file_remove.close()
```

# SVM

Import libraries and open file

```python
#https://medium.com/@vasista/sentiment-analysis-using-svm-338d418e3ff1?fbclid=IwAR0LNbhTKbrd3PuV37rVXbRRMqaVB7cr5nBBDrt6Zq9GFmXyW9ntplbsYs4
# https://deparkes.co.uk/2018/06/18/save-and-load-sci-kit-learn-models/

import time
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix
from sklearn.externals import joblib
import pandas as pd

file1 = pd.read_csv("all_PosNeg.csv")
```

Train and test model

```python
list_tag = []
# Convert Business to 0 and Leisure to 1
for i in range(len(file1)):
    actual_tag = file1.Tags[i].split()
    if (actual_tag[1] == "Leisure"):
        list_tag.append("1")
    elif (actual_tag[1] == "Business"):
        list_tag.append("0")

X_train, X_test, y_train, y_test = train_test_split(file1['Positive_Review'], list_tag, test_size=0.2, random_state=42)

# Create vectors
vectorizer = TfidfVectorizer(min_df = 5, max_df = 0.8, sublinear_tf = True, use_idf = True)
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
# Save vectorizer
joblib.dump(vectorizer, 'training_data.pkl')

# Create SVC
classifier_linear = LinearSVC(random_state=0, tol=1e-5)
t0 = time.time()
# Train model
classifier_linear.fit(X_train, y_train)
t1 = time.time()
# Predict test data
prediction_linear = classifier_linear.predict(X_test)
t2 = time.time()
time_linear_train = t1-t0
time_linear_predict = t2-t1

# Save model
joblib.dump(classifier_linear, 'saved_model.pkl') 

# Time taken for training and testing
print("Training time: %fs; Prediction time: %fs" % (time_linear_train, time_linear_predict))
# Results for Business and Leisure SVM
report = classification_report(y_test, prediction_linear, output_dict=True)
print('Leisure: ', report["1"])
print('Business: ', report['0'])
```

Test load model and vectorizer

```python
clf_load = joblib.load('saved_model.pkl')
vct_load = joblib.load('training_data.pkl')
string = ["Was well located for my business meeting"," Rooms are ok The position is great The metro station is very close"] # Should be business
test = vct_load.transform(string)

prediction_linear = clf_load.predict(test)
print(prediction_linear)
```

Calculate precision, recall, accurancy and f1 score for whole SVM

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html
tn, fp, fn, tp = confusion_matrix(prediction_linear, y_test).ravel()
(tn, fp, fn, tp)
Precision = tp/(tp+fp)
Recall = tp/(tp + fn)
Accurancy = (tp+tn)/(len(y_test))
f1 = 2*((Precision*Recall)/(Precision+Recall))
print("Precision = "+ str(Precision))
print("Recall = "+ str(Recall))
print("accurancy = "+ str(Accurancy))
print("f1 = "+ str(f1))
```
